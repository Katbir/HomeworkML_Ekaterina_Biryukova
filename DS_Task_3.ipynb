1. Вывести все колонки из DataFrame

```python
df = pd.read_csv('TSLA.csv') # открытие файла
df
```

2.Посчитать среднюю цену закрытия акций Tesla, в период с открытия по 10.01.2012

```python
df1 = df[df['Date'] < '2012-01-10'] # вывод строк в период с открытия по 10.01.2012
df1
df1['Adj Close'].mean()  # вывод средней цены закрытия акций Tesla
```

3. Вывести график цены закрытия 

```python
from matplotlib.pyplot import figure

figure(figsize=(15, 3), dpi=120)         # вывод графика цены закрытия
df['Adj Close'].plot()       
```

4. Примените эту формулу нормализации к столбцу Open (используйте метод apply).

```python
min1 = df['Open'].min()   # подсчет максимального значения в столбце Open
max1 = df['Open'].max()   # подсчет минимального значения в столбце Open
min1
```

```python
df['Open'].apply (lambda x: (x-min1)/(max1-min1))  # применяем формулу нормализации к столбцу Open
```

5. Построить график с ценой закрытия и посчитаным SMA 12,15,30

```python
df['Adj Close'].rolling(window = 12).mean().plot() # применение функции SMA для окна 12
```

```python
df['Adj Close'].rolling(window = 15).mean().plot() # применение функции SMA для окна 15
```

```python
df['Adj Close'].rolling(window = 30).mean().plot() # применение функции SMA для окна 30
```
